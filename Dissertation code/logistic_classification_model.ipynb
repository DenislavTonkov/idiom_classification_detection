```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import ast
```

```python
# max lenth of is 45 for idioms 
t=len(x)
r=[]
for i in range(data.shape[0]):
    x = ast.literal_eval(data['all_idiom_5l_each_word'][i])
    r.append(x)
    
```

```python
temp=r
```

```python
for j in range(len(r)):
    for i in range(45-len(r[j])):
        r[j].append(0)
```

```python
data1=pd.concat([pd.DataFrame(r),pd.DataFrame(list(data['label']),columns=['y'])],axis=1)
```

```python
al=[]
for i in range(data1.shape[0]):
    if data1['y'][i]=='i':
        al.append(0)
    elif data1['y'][i]=='l':
        al.append(1)
data1['y']=al      
```

```python
train_idiom=data1[(data1['y']==0)&(data1.index<20000)].reset_index(drop=True)[0:3258]
```

```python
train_literal=data1[(data1['y']==1)&(data1.index<20000)].reset_index(drop=True)[0:3258]
```

```python
train=pd.concat([train_idiom,train_literal],axis=0).reset_index(drop=True)
train
```

```python
test_literal=data1[(data1['y']==1)&(data1.index>20000)].reset_index(drop=True)[0:769]
# 769
```

```python
test_idiom=data1[(data1['y']==0)&(data1.index>20000)].reset_index(drop=True)[0:769]
```

```python
test=pd.concat([test_idiom,test_literal],axis=0).reset_index(drop=True)
test
```

```python
from sklearn.linear_model import LogisticRegression
x_train=train.iloc[:,:45]
y_train=train['y']
x_test=test.iloc[:,:45]
y_test=test['y']
```

```python
model=LogisticRegression(max_iter=300)
model.fit(x_train,y_train)
```

```python
Y_prep=model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y_test,Y_prep)
print (confusion_matrix)
```
